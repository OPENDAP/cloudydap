```python
import os
import numpy as np
from cartopy import config

import cartopy.crs as ccrs
import matplotlib.pyplot as plt
%matplotlib inline


# import mpl_toolkits
# import mpl_toolkits.basemap as Basemap

from netCDF4 import Dataset as netcdf_dataset

opendap_url = 'http://t41m1.opendap.org:8080/opendap/dmrpp_from_aws_cli/\
airs/AIRS.2015.01.01.L3.RetStd_IR001.v6.0.11.0.G15013155825.nc.h5.dmrpp'
DATAFIELD_NAME = 'TotCH4_A_ct'
nc = netcdf_dataset(opendap_url)
data = nc.variables[DATAFIELD_NAME][:,:]
latitude = nc.variables['Latitude'][:]
longitude = nc.variables['Longitude'][:]

# Verify that data is accessible with 
# username & password in ~/.netrc, 
# cookies in ~/.urs_cookies, 
# and remote OPeNDAP server access information in ~/.dodsrc.
print (latitude)
```

If Latitude values are printed correctly, let's plot the grid data on map.

```python
fig = plt.figure(figsize=(12, 8))
ax = plt.axes(projection=ccrs.Mercator())
plt.contourf(longitude, latitude, data, 60, transform=ccrs.Mercator())
ax.coastlines(resolution='10m', color='black')
ax.gridlines()
plt.show()
```
